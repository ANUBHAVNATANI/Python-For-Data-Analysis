# ColorBars in Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Sample Plot

```python
x = np.linspace(0,10,1000)
i = np.sin(x)*np.cos(x[:,np.newaxis])#converting into a 2d array with multiple rows
plt.imshow(i)
plt.colorbar()
```

```python
plt.imshow(i,cmap="gray")
```

### Seeing different cmaps and their properties

1.the jet color map has uneven luminance scale

2.the viridis color map has even luminance scale

3.the cubehelix has also even luminance scale

4.the RdBu(Red blue) is good in showing positive and negative points in preticular

## Some Color Bar customization

```python
s = (np.random.random(i.shape)<0.01)
i[s]=np.random.normal(0,3,np.count_nonzero(s))
plt.figure(figsize=(10,3.5))
plt.subplot(1,2,1)
plt.imshow(i,cmap="RdBu")
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(i,cmap="RdBu")
plt.colorbar(extend="both")
plt.clim(-1,1)
```

in the left panel noise washes out the whole data but in the right the information is pronounced as we have 
set our limits 

### Discrete Color bar

```python
plt.imshow(i,cmap=plt.cm.get_cmap("Blues",6))#bins defined how many shades we want for discrete
plt.colorbar()
plt.clim(-1,1)
```

## Legendry MNIST Example

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits(n_class=6)
fig, ax = plt.subplots(8, 8, figsize=(6, 6))        
for i, axi in enumerate(ax.flat):            
    axi.imshow(digits.images[i], cmap='binary')            
    axi.set(xticks=[], yticks=[])
```

```python
#we can reducde the dimentionality by using manifold
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
projection = iso.fit_transform(digits.data)
plt.scatter(projection[:,0],projection[:,1],lw=0.1,c=digits.target,cmap=plt.cm.get_cmap('cubehelix',6))
plt.colorbar(ticks=range(6),label='digit value')
plt.clim(-0.5,5.5)
```
