# Plot Legends

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Plot

```python
x = np.linspace(0,10,1000)
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),'-r',label="sine")
ax.plot(x,np.cos(x),'-b',label="cosine")
ax.axis("equal")
leg = ax.legend()
```

Changing location of legend

```python
ax.legend(loc="upper left",frameon=False)
fig
```

```python
ax.legend(frameon=False,loc="lower center",ncol=2)#ncol specify number of colums to use to display the legends
fig
```

```python
#we can change how the legend box is shown in the figure
ax.legend(fancybox=True,framealpha=0.9,shadow=True,borderpad=1)
fig
```

```python
#facny function to generate lines
y = np.sin(x[:,np.newaxis]+np.pi*np.arange(0,2,0.5))#np.arange return evenly spaced values with in an intervals
lines = plt.plot(x,y)
plt.legend(lines[:2],["1st","2nd"])
```

```python
#another method
plt.plot(x,y[:,0],label="1st")
plt.plot(x,y[:,1],label="2nd")
plt.plot(x,y[:,2])
plt.legend(framealpha=1,frameon=True)#automatically reject ones without labels
```

### More Complicated Legends

```python
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.rand(100)*1000
w = np.random.rand(100)*100
plt.scatter(x,y,label=None,c=np.log10(z),cmap="viridis",s=w,linewidth=0,alpha=0.5)
plt.axis(aspect="equal")
plt.xlabel("Random")
plt.ylabel("Random")
plt.colorbar(label="Random(z)")
plt.clim(1,3)
for w in [10,30,70]:
    plt.scatter([],[],c="k",alpha=0.3,s=w,label=str(w))
plt.legend(scatterpoints=1,frameon=False,labelspacing=1,title="Random")
plt.title("Random Plot")
```

## Multiple Legends

```python
fig,ax = plt.subplots()
lines = []
styles=["-","--","-.",":"]
x = np.linspace(0,10,1000)
for i in range(4):
    lines += ax.plot(x,np.sin(x-i*np.pi/2),styles[i],color="black")
ax.axis("equal")
ax.legend(lines[:2],["Line A","Line B"],loc="upper right",frameon=False)
#creating custom legends
#Work around to do this
from matplotlib.legend import Legend
leg = Legend(ax,lines[2:],["line C","line D"],loc="lower right",frameon=False)
ax.add_artist(leg)
```
