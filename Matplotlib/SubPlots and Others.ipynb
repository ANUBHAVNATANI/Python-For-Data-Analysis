# Position Of Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Axex in Matplotlib

```python
ax1 = plt.axes()
ax2 = plt.axes([0.6,0.6,0.2,0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],xticklabels=[],ylim=(-1.2, 1.2))       
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],ylim=(-1.2, 1.2)) 
x = np.linspace(0, 10)       
ax1.plot(np.sin(x))       
ax2.plot(np.cos(x))
```

```python
#can be done with subplot easily
plt.subplot(2,1,1)
plt.plot(np.sin(x))
plt.subplot(2,1,2)
plt.plot(np.cos(x))
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4,wspace=0.4)
#hpace is the height space and wsapce is the width space ie done here by 40%
ax = fig.add_subplot(1,2,1)
ax.text(0.5,0.5,str(1),fontsize=16,ha="center")
ax = fig.add_subplot(1,2,2)
ax.text(0.5,0.5,str(2),fontsize=16,ha="center")
```

### In one Go

```python
fig,ax = plt.subplots(2,2,sharex="col",sharey="row")
#sharex and sharey help us to determine the additional property
#by removing sharex and sharey we have removed the inner axis itself
```

```python
for i in range(2):
    for j in range(2):
        ax[i,j].text(0.5,0.5,str(i+j),fontsize=16,ha="center")
fig
```

## More Complex Subplots

```python
grid = plt.GridSpec(2,3,wspace=0.4,hspace=0.3)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:])
plt.subplot(grid[1,:2])
plt.subplot(grid[1,2])
```

## Typically used with histogram

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,3000).T
fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4,hspace=0.4,wspace=0.4)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=main_ax)
x_hist = fig.add_subplot(grid[-1,1:],yticklabels=[],sharex=main_ax)
main_ax.plot(x,y,'ok',markersize=3,alpha=0.2)
x_hist.hist(x,40,histtype="stepfilled",orientation="vertical",color="gray")
x_hist.invert_yaxis()
y_hist.hist(y,40,histtype="stepfilled",orientation="horizontal",color="gray")
y_hist.invert_xaxis()
```
