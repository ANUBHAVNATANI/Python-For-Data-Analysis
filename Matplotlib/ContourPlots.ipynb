# Contour Plots in Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

## Basic Contour Plots

```python
def f(x,y):
    return np.sin(x)+np.cos(y)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x,y)#return coordinate matrices from coordinate vectors
Z = f(X,Y)
plt.contour(X,Y,Z,colors='gray')
```

```python
def f(x,y):
    return np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
```

```python
Z = f(X,Y)
plt.contour(X,Y,Z,colors='gray')
```

```python
plt.contour(X,Y,Z,20,cmap='RdGy')#red grey color map and 20 intervals
```

### Filled Contour

```python
plt.contourf(X,Y,Z,20,cmap="RdGy")
plt.colorbar()
```

For showing more continuous contour we can use plt.imshow

```python
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap="RdGy")
plt.colorbar()
plt.axis(aspect='image')
```

### Combinig Contour Plots and Image Plots

```python
contours = plt.contour(X,Y,Z,3,colors='black')
plt.clabel(contours,inline=True,fontsize=8)
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap="RdGy",alpha=0.5)
plt.colorbar()
```
