# Line Plots in Matlab

```python
#necessary things to do first
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,100)
ax.plot(x,np.sin(x))
```

or we can use the easier interface

```python
plt.plot(x,np.sin(x))
```

To embed multiple in one 

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

Setting the color 

```python
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='g')
```

If no color is specified matplot lib will automatically set the color style by iteslf

We can change the line style also

```python
plt.plot(x,np.sin(x),linestyle='solid')
plt.plot(x,np.sin(x-1),linestyle='dashed')
```

```python
plt.plot(x,np.sin(x),linestyle='-.')
plt.plot(x,np.sin(x-1),linestyle=':')
```

All combined

```python
plt.plot(x,np.sin(x),'-.k')
plt.plot(x,np.sin(x-1),':r')
```

## Adjusting the plot 

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,5)
plt.ylim(-2,6)
```

Reversed

```python
plt.plot(x,np.sin(x))
plt.xlim(5,-1)
plt.ylim(6,-2)
```

### plt.axis command

```python
plt.plot(x,np.sin(x))
plt.axis([-1,5,-2,6])
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal')
#one unit of x equal to one unit of y
```

### Labelling is similar to matlab 

```python
plt.plot(x,np.sin(x))
plt.title("Sin Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

Creating Legends

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis("equal")
plt.legend()
```

### Other way to do all this

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,5),ylim=(-1,1),xlabel="x",ylabel="sin(x)",title="plot")
```
