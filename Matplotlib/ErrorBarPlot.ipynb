# ErrorBar Plot in MatplotLib

```python
#basic imports
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

### Basic Error Bar plot

```python
x = np.linspace(0,10,50)
dy = 0.8 #deviation
y = np.sin(x)+dy*np.random.randn(50) #normal distribution
plt.errorbar(x,y,yerr=dy,fmt='.k') #fmt shows format that black and actual points with dots
```

### Styling error bars

```python
plt.errorbar(x,y,yerr=dy,fmt='o',color='black',ecolor='red',elinewidth=3,capsize=0)
```

## Continuous Error Plot in Matplotlib

Continuous errors plot are not that easy to plot as there is no direct function for them but we can plot it by combinig two different plots

```python
#using gaussian process for example
from sklearn.gaussian_process import GaussianProcessRegressor
xdata = np.array([1,3,5,7,9])
model = lambda x:x*np.sin(x)
ydata = model(xdata)

gp = GaussianProcessRegressor()
gp.fit(xdata[:,np.newaxis],ydata)

xfit = np.linspace(0,10,1000)
yfit , std = gp.predict(xfit[:,np.newaxis],return_std=True)
dyfit = 2*std #how much confidence region here 2sigma ie 95%

#plotting the graph
plt.plot(xdata,ydata,'og')#plots the green points
plt.plot(xfit,yfit,'-',color='red')#plots the red line
plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,color='gray',alpha=0.3)#fills the area with 95% chance to come upper and lower bound fill
plt.xlim(0,10)
```
