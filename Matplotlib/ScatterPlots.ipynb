# Scatter Plots in Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Simple Scatter plot

```python
x = np.linspace(0,10,30) #30 points between 0 to 10
y = np.sin(x)
plt.plot(x,y,'o',color='red')
```

### Many different styles

```python
rng = np.random.RandomState(0) #set the random state
for marker in ['o','v','^','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label="marker='{0}'".format(marker))#random 5 points with in the state defined above
    plt.legend(numpoints=1) #num of points to show in the legend
    plt.xlim(0,1.8)
```

## Plot Combining

```python
plt.plot(x,y,"-ok") # - line,o sactter points,k black
```

### Additional Customization

```python
plt.plot(x,y,'-p',color='green',markersize=12,linewidth=4,markerfacecolor='red',markeredgecolor='blue',markeredgewidth='3')#p here is for pentagon
plt.ylim(-1,1)
```

## Scatter Plots with plt scatter

```python
plt.scatter(x,y)
```

plt scatter let us take control of individual points where as the plt plot let us do that for whole set of points

```python
rng = np.random.RandomState(0)
x = rng.randn(50)#returns a sample of standard normal distribution
y = rng.randn(50)
colors = rng.rand(50)#uniform distribution sample
sizes = 1000*rng.rand(50)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
```

### Implementing it in iris dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0],features[1],alpha=0.4,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
